
# Задание 1

Напишите функцию, которая будет читать файл, лемматизировать текст с помощью pymystem3 и записывать результат в новый файл

```python
from pymystem3 import Mystem

# задаем функцию с двумя аргументами
def function(a, b): 
    f = open(a, encoding = 'utf-8')  # 1 аргумент 
    text = f.read() 
    m = Mystem()
    lemmas = m.lemmatize(text)  # лемматизация текста 
    clean_lemmas = [w for w in lemmas if w.isalpha()]    # избавление от синтаксиса 
    lemmas1=' '.join(clean_lemmas) 
    with open(b, 'w', encoding='utf-8') as f: # 2 аргумент 
        for d in lemmas1:
            f.write(d)
    f.close()

function('Литературные анекдоты.txt', 'lemmas.txt')
```
